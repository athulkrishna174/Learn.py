1. Create a DataFrame by passing a random numpy array,with a datetie index and labelled columns.and display its,
     - Head
     - Tail
     - Index 
     - Columns
     - Values
     - statistics


```python
#please type your answer below this line.
import numpy as np
import pandas as pd
dates = pd.date_range('20210101', periods=15)
df = pd.DataFrame(np.random.randn(15, 26), index=dates, columns=list('ABCDEFGHIJKLMNOPQRSTUVWXYZ'))
print("Head\n", df.head())
print("\nTail\n", df.tail())
print("\nIndex\n", df.index)
print("\nColumns\n", df.columns)
print("\nValues\n", df.values)
print("\nStatistics\n", df.describe(include='all'))
# print("Table\n", df)
```

2. Use the above table and,
      - find transpose of your data.
      - sort the index in the descending order of the date.

```python
#please provide your answer below this line.
print("Transpose\n", df.transpose())
print("\nSorted in Descending Order\n", df.sort_index(ascending=False))
```
